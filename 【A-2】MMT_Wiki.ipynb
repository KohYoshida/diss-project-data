### Wikimedia API https://wikimedia.org/api/rest_v1/

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def wiki_views(language, article):
    my_headers = {"User-Agent": "Your user-agent"}
    endpoint = 'https://wikimedia.org/api/rest_v1'

    # parameters to be included in url
    access = 'all-access'
    agent = 'all-agents'
    language = language
    project = f'{language}.wikipedia' # set the media of Wikimedia, or a language of Wikipedia
    article = article.replace(' ', '_') # any spaces should be replaced by underscores.
    granularity = 'monthly'
    start = '20120101'
    end = '20221231'

    url = f'{endpoint}/metrics/pageviews/per-article/{project}/{access}/{agent}/{article}/{granularity}/{start}/{end}'
    
    req = requests.get(url, headers=my_headers)
    data = req.json()
    df = pd.DataFrame(data['items'])
    return df
```

```python
en_views = wiki_views('en', 'Modern monetary theory')
en_views['timestamp'] = pd.to_datetime(en_views['timestamp'].astype(str).str.slice(0, 6), format='%Y%m')
# drop unnecessary columns
en_views = en_views.drop(['project', 'granularity', 'access', 'agent'], axis=1)
```

```python
"""The title of the Japanese article was changed from 'Modern Monetary Theory' to '現代貨幣理論' on 30th April 2019. 
Therefore, we need to concatenate the two DataFrames. 
The 132 views for '現代貨幣理論' in April 2019 should be ignored 
because they represent a small amount and might be duplicate accesses 
that were redirected from the 'Modern Monetary Theory' page."""

# pageviews for '現代貨幣理論', starting from May 2019 to December 2022
ja_views1 = wiki_views('ja', '現代貨幣理論')
ja_views1['timestamp'] = pd.to_datetime(ja_views1['timestamp'].astype(str).str.slice(0, 6), format='%Y%m')
ja_views1 = ja_views1[ja_views1['timestamp'] >= '2019-05-01']

# pageviews for 'Modern Monetary Theory', starting from from October 2016 to April 2019
ja_views2 = wiki_views('ja', 'Modern Monetary Theory')
ja_views2['timestamp'] = pd.to_datetime(ja_views2['timestamp'].astype(str).str.slice(0, 6), format='%Y%m')
ja_views2 = ja_views2[ja_views2['timestamp'] <= '2019-04-30']

# concatenate these dataframes
ja_views = pd.concat([ja_views1, ja_views2]).sort_values('timestamp')
# drop unnecessary columns
ja_views = ja_views.drop(['project', 'granularity', 'access', 'agent'], axis=1)
```

```python
# merge 'en' and 'ja' dataframes so that we can compare the volumes more easily
pageviews = pd.merge(en_views, ja_views, on='timestamp', how='outer', suffixes=('_en', '_ja'))

# move the column 'timestamp' to the left end
cols = ['timestamp'] + [col for col in pageviews if col != 'timestamp']
pageviews = pageviews[cols]

# change NaN to 0 and float to integer in the column 'views_ja'
pageviews['views_ja'] = pageviews['views_ja'].fillna(0).astype(int)
# change NaN to 'no_article' in the column 'article_ja'
pageviews['article_ja'] = pageviews['article_ja'].fillna('no_article')
# rename 'timestamp' to 'month'
pageviews = pageviews.rename(columns={'timestamp': 'month'})
```

```python
plt.figure(figsize=(10,6))  # specify the size of the figure

# plot the English pageviews
plt.plot(pageviews['month'], pageviews['views_en'], label='English')

# plot the Japanese pageviews
plt.plot(pageviews['month'], pageviews['views_ja'], label='Japanese')

plt.title('Monthly page views comparison on English and Japanese Wikipedia pages of Modern Monetary Theory')  # title of the graph
plt.xlabel('Time')  # label of the x-axis
plt.ylabel('Page views')  # label of the y-axis
plt.legend()  # display the legend
plt.show()  # display the graph
```
